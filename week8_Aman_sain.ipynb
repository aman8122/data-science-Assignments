# 🧠 Single Agent Pipeline Project

## Problem Statement

### The agent should handle:

- General queries → Direct response

# Objective

The objective of this project is to build a **Single-Agent Smart Assistant** that can understand user queries, determine the user's intent, and route the request to the appropriate tool. The assistant is designed to perform mathematical calculations, extract keywords from text, and provide general responses. The project also focuses on implementing conditional routing, tool integration, structured JSON output, and basic error handling while maintaining a modular and scalable architecture.

```python
def calculate(expression):
    try:
        expression = expression.replace("Calculate", "").replace("calculate", "").strip()
        return eval(expression)
    except Exception:
        return "Invalid Expression"
```

```python
def extract_keywords(text):

    stop_words = {
        "the","is","a","an","of","to","and","in","from"
    }

    words = text.lower().split()

    keywords = [word for word in words if word not in stop_words]

    return keywords
```

```python
def respond(query):
    return f"You asked: {query}"
```

```python
import re

def route(query):

    query = query.lower()

    if "calculate" in query:
        return "calculation"

    elif "keyword" in query:
        return "keywords"

    else:
        return "general"
```

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

```python
class SmartAgent:

    def execute(self, query):

        try:

            intent = route(query)

            if intent == "calculation":
                result = calculate(query)

            elif intent == "keywords":
                result = extract_keywords(query)

            else:
                result = respond(query)

            return {
                "type": intent,
                "result": result
            }

        except Exception as e:

            return {
                "type": "error",
                "result": str(e)
            }
```

```python
agent = SmartAgent()
```

```python
queries = [

    "Calculate 20 + 5",

    "Extract keywords from Artificial Intelligence is transforming industries",

    "What is machine learning?"

]

for q in queries:

    print("Query :", q)

    print("Response :", agent.execute(q))

    print("-"*50)
```

```python
while True:

    user_input = input("Enter Query : ")

    if user_input.lower() == "exit":
        break

    print(agent.execute(user_input))
```
